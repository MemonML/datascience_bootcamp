```python
import tkinter as tk
from tkinter import filedialog
from tkinter import messagebox
from PIL import Image, ImageTk
import numpy as np
import tensorflow as tf

# Load a pre-trained model (MNIST CNN)
model = tf.keras.models.load_model('/content/drive/MyDrive/Data_Science_Boot_Camp/week6/day_25/Brain_Tumors.h5')

# Create a Tkinter GUI window
root = tk.Tk()
root.title('MRI Dataset')

# Function to classify the image
def classify_image():
    file_path = filedialog.askopenfilename(filetypes=[("Image files", "*.jpg *.jpeg")])
    if file_path:
        image = Image.open(file_path).convert("L").resize((28, 28))
        image_array = np.array(image) / 255.0
        image_array = image_array.reshape(1, 28, 28, 1)  # Reshape for CNN input
        prediction = model.predict(image_array)
        predicted_class = np.argmax(prediction)

        # Convert PIL Image to PhotoImage
        img_for_display = ImageTk.PhotoImage(image)

        image_label.config(image=img_for_display)
        image_label.image = img_for_display  # Store a reference to avoid garbage collection
        result_label.config(text=f'Predicted Class: {predicted_class}')

open_button = tk.Button(root, text="Open Image", command=classify_image)
image_label = tk.Label(root)
result_label = tk.Label(root, text="Predicted Class: ")

open_button.pack(pady=10)
image_label.pack()
result_label.pack()

root.mainloop()
```
