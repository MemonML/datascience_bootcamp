```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [1, 3, 2, 4, 5]

# Defining figure for 3D plot
fig = plt.figure()
# Defining 3D axis
ax = plt.axes(projection="3d")

# Line plot
ax.plot3D(x, y, z, "red")

# Scatter plot with defined dot color, colormap, and size
dot_color = "green"
dot_size = 50
ax.scatter3D(x, y, z, c=dot_color, cmap="hot", s=dot_size)

# Set labels for the axes
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Set a title for the plot
plt.title('3D Scatter Plot with Labels')

# Displaying figure
plt.show()
```

# Subplot

```python
fig=plt.figure()
# 1st subplot
ax1=fig.add_subplot(1,2,1,projection="3d") # 1---> row, 2---> columns, last one is location
X=np.arange(12)
Y=np.arange(12)
X,Y=np.meshgrid(X,z) # meshgrid
Z=X*2+Y*3

# Add title and labels to the 1st subplot
ax1.set_title('Scatter and Wireframe')
ax1.set_xlabel('X Label')
ax1.set_ylabel('Y Label')
ax1.set_zlabel('Z Label')


ax1.scatter3D(X,Y,Z,c="red")
ax1.plot_wireframe(X,Y,Z)  # wireframe ---> To quickly visulize the 3D object


#2nd subplot
ax2=fig.add_subplot(1,2,2,projection="3d")
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [1, 3, 2, 4, 5]

# Add title and labels to the 1st subplot
ax2.set_title('Scatter')
ax2.set_xlabel('X Label')
ax2.set_ylabel('Y Label')
ax2.set_zlabel('Z Label')

ax2.scatter3D(x,y,z,color="orange")
plt.show()
```

____
Here are some common resampling methods and their characteristics
_____

>>>>Nearest Neighbor (PIL.Image.NEAREST):

>>This method selects the nearest pixel from the original image to determine the pixel value in the resized image. It's the fastest method but can result in pixelation and jagged edges.

>>>>Bilinear (PIL.Image.BILINEAR):

>>This method takes a weighted average of the four nearest pixels in the original image to estimate the pixel value in the resized image. It produces smoother results compared to Nearest Neighbor.

>>>>Bicubic (PIL.Image.BICUBIC):

>>Bicubic interpolation considers 16 neighboring pixels to estimate the pixel value in the resized image. It provides high-quality results with smoother edges and transitions, making it suitable for downsizing images.

>>>>Lanczos (PIL.Image.LANCZOS):

>>Lanczos interpolation uses a sinc function to estimate pixel values. It offers very high-quality results and is often preferred for upsizing images.

>>>>Antialias (PIL.Image.ANTIALIAS):

>>This method is a high-quality resampling method that produces smooth edges and transitions. It can involve different algorithms depending on the image and application.

When using the resample function in PIL, you specify the desired resampling method using constants like PIL.Image.BICUBIC or PIL.Image.LANCZOS. The choice of resampling method depends on factors such as the purpose of the resized image, the content of the image, and the desired trade-off between speed and quality.

>>>>Bilinear and bicubic are two common interpolation methods used in image processing to resize images. They both aim to estimate pixel values at non-integer coordinates when resizing an image. While both methods improve image quality compared to nearest-neighbor interpolation, they have distinct characteristics that make them suitable for different scenarios:

>>>>Bilinear Interpolation:

>>Bilinear interpolation estimates pixel values using a weighted average of the four nearest neighboring pixels. It creates a linear interpolation along both the horizontal and vertical axes. Bilinear interpolation produces smoother results compared to nearest-neighbor interpolation, but it may still introduce some blockiness or artifacts in certain cases. Bilinear interpolation is relatively faster compared to bicubic interpolation.
- Characteristics:
Simplicity and speed
Smooth appearance
May introduce artifacts, especially when resizing images by a large factor
Suitable for quick previews or applications where speed is important


>>>>Bicubic Interpolation:

>>Bicubic interpolation is more sophisticated and accurate than bilinear interpolation. It considers a larger neighborhood of 16 neighboring pixels (4x4 grid) to estimate the pixel value. Bicubic interpolation applies a cubic polynomial along both the horizontal and vertical axes. It results in higher-quality images with smoother edges, transitions, and fewer artifacts compared to bilinear interpolation.
- Characteristics:
Higher image quality
Smoother appearance with fewer artifacts
Slightly slower computational performance compared to bilinear interpolation
Suitable for resizing images when maintaining high quality is a priority


In summary, the main difference between bilinear and bicubic interpolation in terms of image processing is the quality of the resized image. Bicubic interpolation generally produces better-quality results with smoother edges and transitions, making it preferable for situations where image quality is critical. Bilinear interpolation is faster and suitable for cases where speed is more important than achieving the highest quality. The choice between the two methods depends on your specific application and the trade-off between speed and image quality that you are willing to make.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
from sklearn.datasets import load_digits
digits= load_digits() #
```

```python
print("Dataset Shape: ", digits.data.shape)
```

```python
first_image=digits.data[0,:]  # single row which have 64 entries
print(first_image)
```

```python
plt.figure(figsize=(10,4))
plt.imshow(np.reshape(first_image,(8,8)), cmap=plt.cm.gray)
```

**Visualize the data set**

*generating supplotting in a loop*

```python
plt.figure(figsize=(30,7))
for index,(image,label) in enumerate(zip(digits.data[0:5],  digits.target[0:5])):
  plt.subplot(1,5, index+1)
  plt.imshow(np.reshape(image, (8,8)),cmap=plt.cm.gray)
  plt.title('Training %i\n' % label, fontsize=40)
```

**splitting dataset**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test,= train_test_split(digits.data,digits.target, test_size=0.20,random_state=100)
```

```python
print(X_train.shape)
```

**Training & Testing**


```python
knn_classifier = KNeighborsClassifier(n_neighbors=3)  # Using k=5 neighbors
knn_classifier=knn_classifier.fit(X_train, y_train)
```

**Predictions**

```python
y_pred_knn = knn_classifier.predict(X_test)
```

**Model Evaluation**

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
# confusion matrix
cm = metrics.confusion_matrix(y_test, y_pred_knn)
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square=True, cmap='Pastel1')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

```python
print("K-Nearest Neighbors (KNN):")
cm_knn = confusion_matrix(y_test, y_pred_knn)
print("Confusion Matrix")
print(cm_knn)
print("Accuracy: ", accuracy_score(y_test, y_pred_knn))

score_knn = knn_classifier.score(X_test, y_test)
print("Score =", score_knn)
```
